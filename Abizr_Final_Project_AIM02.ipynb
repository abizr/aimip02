#5. masalah missing value

```python
data.isnull().sum()
```

```python
#a. drop kolomyang mengandung missing value lebih dari 50% keseluruhan data
data.drop(columns='company', inplace=True)
```

```python
data.isnull().sum()
```

```python
#b. isi kolom missing value 'country' dan 'children' dari data yang banyak muncul
data['children'].mode()
```

```python
data['country'].mode()
```

```python
df_1 = data
```

```python
df_1 ['children'] = df_1['children'].fillna(df_1['children'].mode()[0])
df_1 ['country'] =  df_1['country'].fillna(df_1['country'].mode()[0])
df_1
```

```python
#c. isi missing value pada kolom agent dengan mean
df_2 = data
```

```python
df_2['agent'].mean()
```

```python
df_2 ['agent'] = df_2 ['agent'].fillna(df_2['agent'].mean())
df_2
```

```python
df_2[['children', 'country', 'agent']]
```

#6.Buat kategpri lead_time: short, medium, long

```python
df_2['lead_time']
```

```python
medianlt= df_2['lead_time'].median()
medianlt
```

```python
df_2['residence_time'] = np.where(df_2['lead_time'] < medianlt, 'short residence', np.where(df_2['lead_time'] < medianlt, 'medium residence', 'long residence'))
df_2
```

```python
df_2['residence_time']
```

#8.Visualisasi

```python
df_2['arrival_date_month'].value_counts()
```

```python
arvl = df_2.groupby(['arrival_date_month']).size()
arvl = arvl.reindex(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
arvl
```

```python
month = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
plt.bar(month, arvl, edgecolor='k', color='#FFADF0')
plt.title('Arrival Date Month')
plt.xlabel('Month')
plt.ylabel('Arrival Data Comaparasion')
plt.show()
```
